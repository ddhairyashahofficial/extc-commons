```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson

# Binomial Distribution
n = 10
p = 0.5
x = np.arange(0, n+1)
binomial_pmf = binom.pmf(x, n, p)

plt.figure()
plt.bar(x, binomial_pmf)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()

# Poisson Distribution
lam = 4
x2 = np.arange(0, 15)
poisson_pmf = poisson.pmf(x2, lam)

plt.figure()
plt.bar(x2, poisson_pmf)
plt.title("Poisson Distribution (lambda=4)")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.show()
```

```python
from scipy.stats import norm, expon

x = np.linspace(-5, 5, 400)

# Normal Distribution
mu, sigma = 0, 1
plt.figure()
plt.plot(x, norm.pdf(x, mu, sigma))
plt.title("Normal Distribution")
plt.xlabel("x")
plt.ylabel("Density")
plt.show()

# Exponential Distribution
x2 = np.linspace(0, 5, 400)
plt.figure()
plt.plot(x2, expon.pdf(x2, scale=1))
plt.title("Exponential Distribution")
plt.xlabel("x")
plt.ylabel("Density")
plt.show()
```

```python
# Normal with different sigma
sigmas = [0.5, 1, 2]

plt.figure()
for s in sigmas:
    plt.plot(x, norm.pdf(x, 0, s), label=f"sigma={s}")
plt.legend()
plt.title("Normal Distribution with Different Sigma")
plt.show()

# Binomial with different p
p_values = [0.2, 0.5, 0.8]
n = 20
x = np.arange(0, n+1)

plt.figure()
for p in p_values:
    plt.plot(x, binom.pmf(x, n, p), marker='o', label=f"p={p}")
plt.legend()
plt.title("Binomial Distribution with Different p")
plt.show()
```

```python
# CLT using dice rolls
sample_size = 30
num_samples = 1000

means = []

for _ in range(num_samples):
    sample = np.random.randint(1, 7, sample_size)
    means.append(np.mean(sample))

plt.figure()
plt.hist(means, bins=30, density=True)
plt.title("CLT Demonstration (Dice Sample Means)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
from scipy.optimize import minimize

# Generate data
data = np.random.exponential(scale=2, size=500)

# Negative log likelihood
def neg_log_likelihood(lam):
    return -np.sum(np.log(lam) - lam * data)

result = minimize(neg_log_likelihood, x0=0.5)
lam_mle = result.x[0]

print("Estimated lambda (MLE):", lam_mle)
print("True lambda approx:", 1/np.mean(data))
```
