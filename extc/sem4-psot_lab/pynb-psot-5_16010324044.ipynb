```python
import math
from scipy import stats

# Given values
x_bar = 3.5
mu0 = 3.25
sigma = 2.61
n = 900
alpha = 0.05

# Z statistic
z = (x_bar - mu0) / (sigma / math.sqrt(n))

# Critical value (two-tailed)
z_crit = stats.norm.ppf(1 - alpha/2)

# p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

z, z_crit, p_value
```

```python
import numpy as np
from scipy import stats

data = [170,174,169,178,180,172,175,177,168,173,
        176,182,171,179,174,176,170,181,172,175]

mu0 = 176
alpha = 0.05

# t-test
t_stat, p_two = stats.ttest_1samp(data, mu0)

# one-sided p-value
p_one = p_two / 2

df = len(data) - 1
t_crit = stats.t.ppf(1-alpha, df)

print("t statistic:", t_stat)
print("t critical:", t_crit)
print("p value:", p_one)
```
