```python
import numpy as np
from scipy import stats

# Sample marks (%) based on sleep duration groups
sleep_less_5 = [55, 60, 58, 62, 57]       # <5 hours
sleep_5_7 = [68, 72, 70, 75, 69]          # 5–7 hours
sleep_more_7 = [78, 82, 80, 85, 79]       # >7 hours

# One-Way ANOVA
f_stat, p_value = stats.f_oneway(sleep_less_5, sleep_5_7, sleep_more_7)

print("F-statistic:", round(f_stat, 2))
print("P-value:", round(p_value, 4))

# Interpretation
if p_value < 0.05:
    print("Significant difference in performance across sleep groups.")
else:
    print("No significant difference across sleep groups.")
```

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Dataset
data = {
    "Marks": [55,60,58, 65,68,70, 78,80,82,
              62,64,66, 72,74,76, 85,88,86,
              50,52,54, 60,62,64, 70,72,74],

    "StudyHours": ["<2","<2","<2","2-4","2-4","2-4",">4",">4",">4",
                   "<2","<2","<2","2-4","2-4","2-4",">4",">4",">4",
                   "<2","<2","<2","2-4","2-4","2-4",">4",">4",">4"],

    "ScreenTime": ["<2","2-5",">5"] * 9
}

df = pd.DataFrame(data)

# Convert to category
df["StudyHours"] = df["StudyHours"].astype("category")
df["ScreenTime"] = df["ScreenTime"].astype("category")

# Two-Way ANOVA with interaction
model = smf.ols('Marks ~ C(StudyHours) + C(ScreenTime) + C(StudyHours):C(ScreenTime)', data=df).fit()

anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

data = {
    "WPM": [28,30,29, 34,36,35, 40,42,41,
            26,27,25, 31,33,32, 37,38,39],

    "Gender": ["Male"]*9 + ["Female"]*9,

    "Usage": ["<2","2-5",">5"] * 6
}

df = pd.DataFrame(data)

df["Gender"] = df["Gender"].astype("category")
df["Usage"] = df["Usage"].astype("category")

model = smf.ols('WPM ~ C(Gender) + C(Usage) + C(Gender):C(Usage)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)
```
